```python
age = input("What is your current age?")
```

```python
rest_of_life_yr = 90-int(age)
rest_of_life_mn = rest_of_life_yr*12
rest_of_life_wk = rest_of_life_yr*52
rest_of_life_dy = rest_of_life_yr*365

print(f"You have {rest_of_life_dy} days, {rest_of_life_wk} weeks, and {rest_of_life_mn} months left.")
```

Tip Calculator

```python
print("Welcome to the tip calculator.")
total_amount = input("What was the total bill? ")
tip_per = input("What % of tip would you like to give? ")
ppl = input("How many people to split the bill? ")
```

```python
split_val = round((float(total_amount) * (1+float(tip_per)/100)) / int(ppl),2)
format_split_val = "{:.2f}".format(split_val)

print(f"Each person should pay {format_split_val}")
```
